## Application Programming Interfaces

In WK09 we saw how to get data/info from arbitrary web pages by requesting their `html`/`xml` code and parsing out any content of interest.

This week we’re going to explore another way of requesting data from websites.

We’re still making requests, but this time instead of requesting web page files from servers, we’re going to request data from API services. APIs, or _application programming interfaces_, are services that make data accessible through code.

For example: we can query the NYT’s database for articles about the environment from the last ten years, or ask a public meteorological service for the current temperature in Manhattan. We’re still making requests to web addresses, but instead of getting web sites as the response we (most likely) get `JSON` data.

There are no standards for the kind of data that is returned or its format, as each API has its own set of conventions and ways it structures data.

### Meteorological Data

Let’s start with a simpler service: weather and geolocation.

By passing different parameters to different URL endpoints we can query the same service for different types and quantities of data.

The service is [Open Meteo](https://open-meteo.com/) and their API is described [HERE](https://open-meteo.com/en/docs).

This is a good service to start with because we basically fill out a form and in the end we get an URL that we can use in our code.

#### Geolocation

There are many, many services that this API provides, but we can start simple.

Let's get latitude and longitude for New York and a second city of our choice.

If we look [here](https://open-meteo.com/en/docs/geocoding-api), for documentation about the Geocoding API, and fill out the form at the top:

We can create a request URL with the following format:

`https://geocoding-api.open-meteo.com/v1/search?name=New+York&count=10&language=en&format=json&countryCode=US`

which we can use in Python to get the latitude and longitude of New York City.

```python
REQ_URL = "https://geocoding-api.open-meteo.com/v1/search?name=New+York&count=10&language=en&format=json&countryCode=US"
data_response = requests.get(REQ_URL)

# Get json response (not html, not text)
data = data_response.json()

# This is a python object/dictionary that we can index to get the data we're interested in
display(data)
```

#### Weather

Now, let's use the weather service to get hourly temperatures for both cities for the month of April.

If we look at the Historical Weather API, [here](https://open-meteo.com/en/docs/historical-weather-api), we'll see a similar form that we can fill out to build our request URL:

<img src="./imgs/01-weather_00.jpg" width="600px" />

We can create a request URL with the following format:

`https://archive-api.open-meteo.com/v1/archive?latitude=40.71427&longitude=-74.00597&start_date=2026-04-01&end_date=2026-04-18&hourly=temperature_2m`

which we can use in Python to get historical weather data for New York City.

```python
# We can use some variables to make the URL easier to build/change

START = "2026-04-01"
END = "2026-04-18"
LAT = "40.71427"
LON = "-74.00597"

REQ_URL = f"https://archive-api.open-meteo.com/v1/archive?latitude={LAT}&longitude={LON}&start_date={START}&end_date={END}&hourly=temperature_2m"
data_response = requests.get(REQ_URL)

# Get json response (not html, not text)
ny_data = data_response.json()

display(ny_data)
```

## API Keys

Some apis are capable of returning so much data (and/or private data) that its maintainers have to add some type of authentication to make sure they know who is making the requests and how often.

Such is the case for the APIs of the New York Times and the Smithsonian collection.

We can get free api keys for these services from [https://developer.nytimes.com/apis](https://developer.nytimes.com/apis) and [https://api.data.gov/signup/](https://api.data.gov/signup/), and include them in our requests as a way to identify who we are. API keys are like passwords and should be treated as such: no sharing, no posting online, no committing on GitHub.

Let’s create keys for either or both services and see how to use them below.

- [NY Times API Documentation](https://developer.nytimes.com/docs/articlesearch-product/1/overview)

- [Smithsonian API Documentation](https://edan.si.edu/openaccess/apidocs/)

## NYT API

We can create a query for the NYT API using the following URL format:

`https://api.nytimes.com/svc/search/v2/articlesearch.json?q={query}&api-key={key}`

```python
QUERY = "climate change"

REQ_URL = f"https://api.nytimes.com/svc/search/v2/articlesearch.json?q={QUERY}&api-key={NYT_KEY}"
data_response = requests.get(REQ_URL)

nyt_data = data_response.json()

display(nyt_data)
```

### Comparing Images

We can pass a list of opened images to the LLM and ask for it to describe common themes in them.

```python
imgs = []
for cnt in range(3):
  imgs.append(PImage.open(f"imgs/GDTM-{cnt}.jpg"))

response = client.models.generate_content(
  model="gemini-3-flash-preview",
  contents=[
    imgs,
    "What do these images have in common? Respond in plain text, not markdown.",
  ]
)

response_data = response.to_json_dict()
```
